```python
#1 celda
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Dense, RepeatVector, TimeDistributed
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

plt.style.use('ggplot')
```

```python
#2.5 Celda
try:
    df_con_indice = df.set_index('ingresado_local')
    df_con_indice.index = pd.to_datetime(df_con_indice.index, errors='coerce')
    df_con_indice = df_con_indice.loc[~pd.isna(df_con_indice.index)].sort_index()

except KeyError:
    print("ERROR: La columna 'ingresado_local' no existe. Revisa el df.head() de la celda anterior.")

    raise

ID_CUARTO_NORMAL = 4
df_cuarto_1 = df_con_indice[df_con_indice['cuarto_id'] == ID_CUARTO_NORMAL].copy()

if df_cuarto_1.empty:
    print(f"ALERTA: No se encontraron datos para el cuarto_frio_id = {ID_CUARTO_NORMAL}.")
    print("Revisa los IDs disponibles en tus datos.")

fecha_fin = df_cuarto_1.index.max()
fecha_ini = max(df_cuarto_1.index.min(), fecha_fin - pd.Timedelta(minutes=60))
df_normal = df_cuarto_1.loc[(df_cuarto_1.index >= fecha_ini) & (df_cuarto_1.index <= fecha_fin)].copy()


if df_normal.empty:
    print("ALERTA: El filtro de fechas 'normales' no encontró datos. Ajusta las fechas.")

features_a_usar = ['temperatura_c', 'humedad_pct']
df_entrenamiento = df_normal[features_a_usar]

print(f"Datos listos para entrenar: {len(df_entrenamiento)} lecturas (Cuarto {ID_CUARTO_NORMAL})")
if not df_entrenamiento.empty:
    df_entrenamiento['temperatura_c'].plot(title=f"Temperatura 'Normal' - Cuarto {ID_CUARTO_NORMAL}", figsize=(12,5))
    plt.show()
```

```python
# CELDA 3 

# 1. Escalar los datos:
scaler = MinMaxScaler(feature_range=(0, 1))

# ¡¡IMPORTANTE!! Asegúrate de que usa 'df_entrenamiento'
# (que ya tiene las 2 columnas ['temperatura_c', 'humedad_pct'])
datos_normales_scaled = scaler.fit_transform(df_entrenamiento)

# 2. Crear secuencias:
TIME_STEPS = 6 

def crear_secuencias(datos, time_steps):
    Xs = []
    for i in range(len(datos) - time_steps):
        Xs.append(datos[i:(i + time_steps)])
    return np.array(Xs)

X_train = crear_secuencias(datos_normales_scaled, TIME_STEPS)
print(f"Forma de los datos de entrenamiento (Secuencias, Pasos, Features): {X_train.shape}")
```

```python
#4 celda
timesteps = X_train.shape[1]
features = X_train.shape[2]  

inputs = Input(shape=(timesteps, features))
encoding = LSTM(16, activation='relu')(inputs)

decoding = RepeatVector(timesteps)(encoding)
decoding = LSTM(16, activation='relu', return_sequences=True)(decoding)

outputs = TimeDistributed(Dense(features))(decoding)


model = Model(inputs=inputs, outputs=outputs)
model.compile(optimizer='adam', loss='mae') 
model.summary()
```

```python
#5 celda
history = model.fit(
    X_train, 
    X_train,  
    epochs=30,
    batch_size=32,
    validation_split=0.1, 
    shuffle=False
)


plt.plot(history.history['loss'], label='Error (entrenamiento)')
plt.plot(history.history['val_loss'], label='Error (validación)')
plt.legend()
plt.title("El modelo está aprendiendo")
plt.show()
```

```python
#6 celda
X_train_pred = model.predict(X_train)

train_mae_loss = np.mean(np.abs(X_train_pred - X_train), axis=1)

plt.hist(train_mae_loss, bins=50)
plt.title("Distribución del Error de Reconstrucción (Datos Normales)")
plt.show()

umbral = np.quantile(train_mae_loss, 0.95)
print(f"UMBRAL DE ANOMALÍA ESTABLECIDO EN: {umbral}")
```

```python
# CELDA 7 
df_anomalo_real = df_cuarto_1.loc['2025-10-27 00:00:00':'2025-10-28 12:00:00'].copy() 

features_a_usar = ['temperatura_c', 'humedad_pct'] 
df_prueba = df_anomalo_real[features_a_usar]

if df_prueba.empty:
    print("ALERTA: El filtro de fechas 'anómalas' no encontró datos. Ajusta las fechas.")
else:
    print(f"Probando con {len(df_prueba)} lecturas anómalas...")
    df_prueba['temperatura_c'].plot(title="Datos de Prueba (Anómalos)")
    plt.show()

    datos_test_scaled = scaler.transform(df_prueba) 
    X_test = crear_secuencias(datos_test_scaled, TIME_STEPS) 

    if X_test.shape[0] > 0:
        test_pred = model.predict(X_test)
        test_mae_loss = np.mean(np.abs(test_pred - X_test), axis=1)

        indice_grafica = df_prueba.index[TIME_STEPS:]
        
        plt.plot(indice_grafica, test_mae_loss, label='Error de prueba')
        plt.axhline(y=umbral, color='r', linestyle='--', label='Umbral de Anomalía')
        plt.legend()
        plt.title("¡DETECCIÓN DE ANOMALÍAS REALES!")
        plt.show()
    else:
        print("No se generaron secuencias de prueba. Asegúrate de tener suficientes datos anómalos (más que TIME_STEPS).")
```
